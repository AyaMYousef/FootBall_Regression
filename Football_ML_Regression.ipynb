```python
 # Dropping Outliers

def discard_outliers(dataf):
  Q1 = dataf.quantile(0.25)
  Q3 = dataf.quantile(0.75)
  IQR = Q3 -Q1

  normal_point = ~((dataf <(Q1 -1.5*IQR)) | (dataf > (Q3 +1.5 * IQR))).any(axis=1)
  removed_outliers_df = dataf[normal_point].reset_index(drop=True)

  return removed_outliers_df
```

```python
df = discard_outliers(df)
df
```

# Regression Models:


-- regression models such as (linear regression, knn, svr, random
forest, decision tree, adaboost, xgboost)

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
```

```python

from sklearn.metrics import r2_score, mean_absolute_error,  mean_squared_error, classification_report
```

```python


models = {
    'LinearRegressor' : LinearRegression(),
    'KNN' : KNeighborsRegressor(),
    'SVM' : SVR(),
    'DecisionTree' : DecisionTreeRegressor(),
    'RandomForest' : RandomForestRegressor(),
    'ADABoost'     : AdaBoostRegressor(),
    'XGBoost'      : XGBRegressor()
}
```

```python
def evaluate_models(actual, predicted):
  score = r2_score(actual, predicted)
  mean_absolute_err = mean_absolute_error(actual, predicted)
  mean_squared_err  = mean_squared_error(actual, predicted)
  root_mean_squared_err = np.sqrt(mean_squared_err)
  return 'Score = ', score, 'Mean_AbsoluteErr =', mean_absolute_err, 'Mean_Squared_Err =', mean_squared_err, 'Root MeanSquaredErr', root_mean_squared_err
```

```python
for i in range(len(models.values())):
  model = list(models.values())[i]
  model_name = list((models.keys()))[i]
  model.fit(x_train, y_train)

  y_pred_test = model.predict(x_test)
  y_pred_train = model.predict(x_train)

  print(model_name)
  print('Test', model_name)
  print(evaluate_models(y_test, y_pred_test))
  print('Train', model_name)
  print(evaluate_models(y_train, y_pred_train))
  print('-----------------------------------------------------------------------------------------------------------------------------')
```

-- Based on the Evaluating criteris :
   

    *  R2-score : Evaluates the variance of the labels.(Higher is better)
    *  MAE, MSE, RMSE :  indicates the average value of error in predicting (Lower is better)

* Based on the results above the best two models are XGBOOSt and RandomForest



```python
# Grid Search for both models
# RandomForest
from sklearn.model_selection import GridSearchCV
params = {
    'n_estimators' :[50,100,150],
    'min_samples_split' : [2,3,4],
    'min_samples_leaf' : [1,2,3]
}
model = RandomForestRegressor()
grid_search = GridSearchCV(model, params)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
model_modified = RandomForestRegressor(min_samples_leaf= 1, min_samples_split= 4, n_estimators= 150)
model_modified.fit(x_train, y_train)
```

```python
y_pred_test = model_modified.predict(x_test)
```

```python
print('Test RandomForest')
print(r2_score(y_test, y_pred_test))
print(mean_absolute_error(y_test, y_pred_test))
print(mean_squared_error(y_test, y_pred_test))
print(np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

```python
y_pred_train = model_modified.predict(x_train)
```

```python
print('Train RandomForest')
print(r2_score(y_train, y_pred_train))
print(mean_absolute_error(y_train, y_pred_train))
print(mean_squared_error(y_train, y_pred_train))
print(np.sqrt(mean_squared_error(y_train, y_pred_train)))
```

```python
#XGBOOST
xgb = XGBRegressor()
params= { 'learning_rate': [0.03, 0.005, 1],
          'max_depth': [5, 6, 7],
          'n_estimators': [500,1000]}

xgb_grid = GridSearchCV(xgb,params)

xgb_grid.fit(x_train, y_train)

print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
xgb_modified = XGBRegressor(learning_rate = 0.03, max_depth= 5, n_estimators =1000)
xgb_modified.fit(x_train, y_train)
```

```python
y_pred_test = xgb_modified.predict(x_test)
```

```python
print('Test XGBoost')
print(r2_score(y_test, y_pred_test))
print(mean_absolute_error(y_test, y_pred_test))
print(mean_squared_error(y_test, y_pred_test))
print(np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

```python
y_pred_train = xgb_modified.predict(x_train)
```

```python
print('Train XGBoost')
print(r2_score(y_train, y_pred_train))
print(mean_absolute_error(y_train, y_pred_train))
print(mean_squared_error(y_train, y_pred_train))
print(np.sqrt(mean_squared_error(y_train, y_pred_train)))
```
